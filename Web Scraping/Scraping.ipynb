```python
instituciones = pd.Series()
```

```python
institucion_list = []
```

```python
def parsear_institucion(id_institucion):
    urls = "https://portal.siges.sv"+id_institucion
    browser.get(urls)
    browser.implicitly_wait(10)
    html = browser.page_source
    soup = bs(html, 'lxml')
    
    
    nombre = soup.find('span',{'id':'form:input_nombre'}).text
    print('\n Nombre: ' + nombre)
    
    codigo = soup.find('span',{'id':'form:input_codigo'}).text
    print('\n Codigo: ' + codigo)
    
    etiqueta_direccion = soup.find('label', string='Dirección:')
    direccion = etiqueta_direccion.find_previous().get_text(strip=True).split(':')[1]
    print('\n Direccion; ' + direccion)
    
    telefono = soup.find('span',{'id':'form:input_telefono'}).text
    print('\n Telefono: ' + telefono)
    
    instituciones['Nombre'] = nombre
    instituciones['Codigo'] = codigo
    instituciones['Direccion'] = direccion
    instituciones['Telefono'] = telefono

    institucion_info ={
        'Nombre:' : nombre,
        'Codigo:' : codigo,
        'Direccion' : direccion,
        'Telefono' : telefono
    }

    institucion_list.append(institucion_info)

    df_instituciones = pd.DataFrame(instituciones)
    
    return(df_instituciones.T)
    
```

```python
df_instituciones = parsear_institucion(ids_instituciones.iloc[0].id)

contador = 0
for elemento in id_instituciones:
    contador += 1

for i in range(1,contador):
    df_instituciones = pd.concat([df_instituciones,parsear_institucion(id_instituciones[i])])
    time.sleep(random.randint(1,3))
```
